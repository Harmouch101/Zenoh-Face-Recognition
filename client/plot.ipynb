```python
import plotly as pt
import plotly.graph_objs as go
import numpy as np

# open with read then store in numpy *how I want to do it*
y_axis = []
avg = 0
samples = 0
#with open('rate.txt') as f:
with open('delay.txt') as f:
    #Content_list is the list that contains the read lines.     
    for line in f:
        y_axis.append(float(line))
        avg += float(line)
        samples +=1
    avg /= samples
print(avg)
avg_arr = np.full((1, samples),avg)
x_axis = np.arange(samples)
#plot the model for prediction
layout = go.Layout(title =  dict(text ='Delay',font =dict(family='Sherif',size=18,color = 'white')),
    xaxis=dict(title="samples", zeroline=False,showgrid = True,gridcolor='gray'),
    yaxis=dict(title="delay",zeroline = False,showgrid = True, tickangle = 60,gridcolor='gray'),
    paper_bgcolor='#57565e',plot_bgcolor='rgba(0,0,0,0)',font = dict(color = 'lightyellow'))
fig = go.FigureWidget(data=[
    go.Scatter(name="ts_delay", x=x_axis,y=y_axis,line=dict(color="aqua"),mode="lines"),
    go.Scatter(name="mean", x=x_axis,y=avg_arr[0],line=dict(color="lightgreen"),mode="lines")],layout=layout)
# Change the bar mode
fig.layout.update(barmode='group')
fig.show()
```
